```python
# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

## Анализ исторических данных

```python
two_sigmas = 2 ### интервал для аномалий
window = 30  ### окно скользящего среднего
```

```python
### анализ 1-ого города
def analyze_city(city_df: pd.DataFrame, ### должны подавать только отсортированные по дате данные (для rolling)
                 window: int = 30):
    city_name = city_df.loc[0,'city']

    ### скользящее среднее и стандартное отклонение по всему городу
    city_df["rolling_mean"] = city_df["temperature"].rolling(window=window).mean()
    city_df["rolling_std"] = city_df["temperature"].rolling(window=window).std()

    ### аномалии по скользящему среднему
    city_df["is_anomaly"] = city_df.apply(lambda row:
                                          row['temperature'] > row["rolling_mean"] + two_sigmas * row["rolling_std"]
                                          or
                                          row['temperature'] < row["rolling_mean"] - two_sigmas * row["rolling_std"],
                                          axis=1)

    ### avg, min, max температуры по городу за все время
    avg_temp = city_df["temperature"].mean()
    min_temp = city_df["temperature"].min()
    max_temp = city_df["temperature"].max()

    ### статистика по сезонам
    city_season_df = city_df.groupby("season")["temperature"].agg(season_avg_temp="mean",
                                                                  season_std_temp="std").reset_index()

    ### тренд
    X = pd.DataFrame((city_df["timestamp"] - city_df["timestamp"].min()).dt.days)
    y = city_df["temperature"]

    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    trend = "положительный" if model.coef_[0] > 0 else "отрицательный"

    return {
        "city": city_name,
        "avg_temp": avg_temp,
        "min_temp": min_temp,
        "max_temp": max_temp,
        "seasonal_profile": city_season_df,
        "trend": trend,
        "anomalies": city_df[city_df.is_anomaly == True]
    }

def inner_analyze_city(input_df: pd.DataFrame,
                       city: str):
    city_df = input_df[input_df["city"] == city].reset_index(drop=True)
    return analyze_city(city_df)

### вывод результатов
def display_results(results):
    for i,result in enumerate(results):

        print(f"Город {i+1}: {result['city']}")
        print(f"Средняя температура: {result['avg_temp']:.2f}°C")
        print(f"Мин. температура: {result['min_temp']:.2f}°C")
        print(f"Макс. температура: {result['max_temp']:.2f}°C")
        print(f"Тренд температуры: {result['trend']}")

        print("\nПрофиль сезона:")
        display(result["seasonal_profile"])

        print("\nАномальные дни:")
        display(result["anomalies"])

        print("-" * 100)
```

### Без распараллеливания


```python
### последовательный анализ всех городов (без распараллеливания)
def sequential_analysis(input_df: pd.DataFrame):
  results = []
  cities = input_df["city"].unique()

  for city in cities:
      city_df = input_df[input_df["city"] == city].reset_index(drop = True)
      result = analyze_city(city_df)
      results.append(result)

  return results
```

### С параллеливанием (многопоточность)

**С многопоточностью расчеты не стали быстрее**  
**Что логично, так как все равно работает 1 поток в моменте, а расчеты - CPU-bound**


```python
### многопоточный анализ всех городов
def multithreading_analysis(input_df: pd.DataFrame,
                            max_workers: int = 4):
  results = []
  cities = input_df["city"].unique()

  ### ThreadPoolExecutor для распараллеливания (многопоточности)
  with ThreadPoolExecutor(max_workers=max_workers) as executor:
      futures = [executor.submit(inner_analyze_city, input_df, city) for city in cities]
      for future in futures:
          results.append(future.result())

  return results
```

```python
results = multithreading_analysis(df, max_workers=4)
display_results(results)
```

```python
times_passed = []
for _ in tqdm(range(50)):
  start_time = time.time()
  results = multithreading_analysis(df, max_workers=4)
  time_passed = time.time() - start_time
  times_passed.append(time_passed)

median = np.percentile(times_passed, 50)
lower_bound = np.percentile(times_passed, 5)
upper_bound = np.percentile(times_passed, 95)

print(f'\nПроцесс расчетов занимает, в среднем, {median:.2f} секунд.')
print(f'5% нижняя граница - {lower_bound:.2f}')
print(f'95% верхняя граница - {upper_bound:.2f}')
```

### С распараллеливанием (многопроцессность)

**С многопроцессностью расчеты также не стали быстрее**  
**Подозреваю, что дело в маленьком объеме расчетов, в сравнении с которыми накладные расходы весят много и не результируют в итоговом выигрыше подхода**


```python
### многопроцессный анализ всех городов
def multiprocessing_analysis(input_df: pd.DataFrame,
                            max_workers: int = 4):
  results = []
  cities = input_df["city"].unique()
  city_data = {
        city: input_df[input_df["city"] == city].reset_index(drop=True)
        for city in cities
    }

  ### ProcessPoolExecutor для распараллеливания (многопроцессность)
  with ProcessPoolExecutor(max_workers=max_workers) as executor:
      futures = [executor.submit(analyze_city, data) for data in city_data.values()]
      for future in futures:
          results.append(future.result())

  return results
```

## Мониторинг текущей температуры

```python
### ключи, пути, параметры
API_KEY = '' ### нужно вставить в ключ
LAT_LON_URL = 'http://api.openweathermap.org/geo/1.0/direct'
TEMP_URL = "https://api.openweathermap.org/data/2.5/weather"
season = 'winter'
cities = df['city'].unique()
```

```python
### проверяем на аномальность текущую температуру с профилем сезона
def anomality_check(current_temp: float,
                    seasonal_profile: pd.DataFrame,
                    season: str):
  '''
  Определяем, является ли текущая температура аномальной.
  '''
  season_df = seasonal_profile[seasonal_profile["season"] == season]
  mean_temp = season_df["season_avg_temp"].values[0]
  std_temp = season_df["season_std_temp"].values[0]

  return current_temp < mean_temp - two_sigmas * std_temp or current_temp > mean_temp + two_sigmas * std_temp
```

### Синхронный (последовательный) подход


```python
### хотим делать все запросы внутри одной сессии, чтобы сократить накладные расходы на переподключения

def get_city_lat_lon(city: str,
                     session: requests.Session):
  '''
  Получаем широту и долготу города по его названию
  '''
  params = {
      "q": city,
      "appid": API_KEY,
  }
  response = session.get(LAT_LON_URL, params=params)
  if response.status_code == 200:
    result = response.json()
    return result[0]['lat'],result[0]['lon']
  else:
    raise Exception(f"Ошибка получения данных: {response.status_code}, {response.text}")

def get_city_current_temperature(lat: float,
                                 lon: float,
                                 session: requests.Session):
  '''
  Получаем текущую температуру по широте и долготе города
  '''
  params = {
      "lat": lat,
      "lon": lon,
      "appid": API_KEY,
      "units": "metric"  ### переводим в градусы Цельсия
  }
  response = session.get(TEMP_URL, params=params)
  if response.status_code == 200:
    result = response.json()
    return result["main"]["temp"]
  else:
    raise Exception(f"Ошибка получения данных: {response.status_code}, {response.text}")
```

```python
### получаем для каждого города текущую температуру через OpenWeatherMap API.
### Проверяем на аномальность текущую температуру с профилем сезона

start_time = time.time()

with requests.Session() as session:
  for i,city in enumerate(cities):
    lat, lon = get_city_lat_lon(city, session)
    current_temp = get_city_current_temperature(lat, lon, session)
    result = inner_analyze_city(df, city)
    seasonal_profile = result['seasonal_profile']
    anomaly = 'не нормальная' if anomality_check(current_temp, seasonal_profile, season) else 'нормальная'
    print(f"{i+1}) Текущая температура в {city}: {current_temp:.2f}°C - {anomaly}")

time_passed = time.time() - start_time
print(f'Процесс расчетов занял {time_passed:.2f} секунд.')
```

### Асинхронный подход

**Асинхронный подход дает ощутимый прирост в скорости**  

```python
### передаем активную сессию в асинхронные функции
### То есть, внутри сессий нам остается делать только сами запросы

async def async_get_city_lat_lon(city: str,
                                session: aiohttp.ClientSession):
  """
  Асинхронно получаем широту и долготу города по его названию
  """
  params = {
      "q": city,
      "appid": API_KEY,
  }
  async with session.get(LAT_LON_URL, params=params) as response:
      if response.status == 200:
        result = await response.json()
        return result[0]['lat'], result[0]['lon']
      else:
        raise Exception(f"Ошибка получения данных: {response.status}, {await response.text()}")

async def async_get_city_current_temperature(lat: float,
                                             lon: float,
                                             session: aiohttp.ClientSession):
  """
  Асинхронно получаем текущую температуру по широте и долготе города
  """
  params = {
      "lat": lat,
      "lon": lon,
      "appid": API_KEY,
      "units": "metric",  # Переводим в градусы Цельсия
  }
  async with session.get(TEMP_URL, params=params) as response:
      if response.status == 200:
        result = await response.json()
        return result["main"]["temp"]
      else:
        raise Exception(f"Ошибка получения данных: {response.status}, {await response.text()}")
```

```python
### делаем одну сессию на все таски, чтобы сократить накладные расходы
### асинхронно по каждому городу
### получаем широту и долготу
### получаем текущую температуру
### проверяем текущую температуру на аномальность

async def city_task(i : int,
                    city: str,
                    input_df: pd.DataFrame,
                    session: aiohttp.ClientSession):
  try:
      lat, lon = await async_get_city_lat_lon(city, session)
      current_temp = await async_get_city_current_temperature(lat, lon, session)

      result = inner_analyze_city(input_df, city)
      seasonal_profile = result['seasonal_profile']

      anomaly = 'не нормальная' if anomality_check(current_temp, seasonal_profile, season) else 'нормальная'

      print(f"{i+1}) Текущая температура в {city}: {current_temp:.2f}°C - {anomaly}")
  except Exception as e:
      print(f"Ошибка обработки для города {city}: {e}")
```

```python
start_time = time.time()

async def main(cities: list,
               input_df: pd.DataFrame):
  """
  Основная асинхронная функция для обработки всех городов параллельно.
  """
  async with aiohttp.ClientSession() as session:
      tasks = [city_task(i, city, input_df, session) for i,city in enumerate(cities)]
      await asyncio.gather(*tasks)

asyncio.run(main(cities, df))

time_passed = time.time() - start_time
print(f'Процесс расчетов занял {time_passed:.2f} секунд.')
```
